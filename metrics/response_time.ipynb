```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
data = pd.read_csv('response_time.csv')

# Extract the number of nodes
nodes = data.columns[1:].astype(int).values

# Extract the response times for each attempt
attempt1 = data.iloc[0, 1:].astype(int).values
attempt2 = data.iloc[1, 1:].astype(int).values
attempt3 = data.iloc[2, 1:].astype(int).values
attempt4 = data.iloc[3, 1:].astype(int).values
attempt5 = data.iloc[4, 1:].astype(int).values

# Create a scatter plot
plt.figure(figsize=(10, 6))

plt.scatter(nodes, attempt1)
plt.scatter(nodes, attempt2)
plt.scatter(nodes, attempt3)
plt.scatter(nodes, attempt4)
plt.scatter(nodes, attempt5)

# Add a trend line (average response time per node count)
all_attempts = np.vstack([attempt1, attempt2, attempt3, attempt4, attempt5])
avg_times = np.mean(all_attempts, axis=0)
plt.plot(nodes, avg_times, 'k--', label='Average', linewidth=2)

# Add labels and title
plt.ylim(top=1500)  # adjust the top leaving bottom unchanged
plt.ylim(bottom=500)  # adjust the bottom leaving top unchanged
plt.xlabel('Number of Nodes')
plt.ylabel('Response Time (ms)')
plt.title('Response Time vs Number of Nodes')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Set x-axis to show only integer values
plt.xticks(nodes)

# Show the plot
plt.tight_layout()
plt.show()
```
